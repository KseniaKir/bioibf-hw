```python
import pandas as pd 
tsv =  'BRCA_pam50.tsv'
df = pd.read_csv(tsv, sep='\t', index_col=0)
df.head()
```

```python
genes = ['ESR1', 'PGR', 'ERBB2', 'MKI67', 'Subtype']
TCGA = df.index.to_list()
d = {'TCGA': TCGA}
df_favs = pd.DataFrame(data=d)
for i in genes:
    gene_column = pd.Series(df.loc[:, i])
    gene_column = gene_column.map(lambda x: str(x)).to_list()
    df_favs[i] = gene_column
df_favs = df_favs.set_index('TCGA')
```
