```python
#4
from scipy.stats  import t, norm
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

def variab(a):
    rv = t(1)
    arr = []
    for i in range (1000):
        sample = rv.rvs(size=a)
        s = np.mean(sample)
        arr.append(s)
    return arr

rv2 = t(1)
nums = [0]*1000
x=-5
for i in range (1000):
    nums[i]=x+0.01
    x+=0.01
df_d = pd.DataFrame({"x": nums})
df_d["PDF"] = rv2.pdf(df_d["x"])

def plots(a):
    sns.histplot(x=a, stat="density")
    sns.lineplot(data=df_d, x="x", y="PDF")
    plt.xlim(-5, 5)
    return plt.tight_layout()

arr2 = variab(2)
plots(arr2)
```

```python
#5
from scipy.stats  import t, norm
import pandas as pd
import seaborn as sns
import math
import numpy as np
import matplotlib.pyplot as plt

def variab(a):
    rv = t(3)
    arr = []
    for i in range (1000):
        sample = rv.rvs(size=a)
        s = np.mean(sample)
        arr.append(s)
    return arr

def expect(a, n):
    prob = 1/n
    summ = 0
    for i in range(0, n):
        summ += (a[i] * prob)
    return summ

arr2 = variab(2)
E = expect (arr2, 1000)

rv2 = norm(E, math.sqrt(3))
nums = [0]*1000
x=-5
for i in range (1000):
    nums[i]=x+0.01
    x+=0.01
arr2_2 = pd.DataFrame({"x": nums})
arr2_2["PDF"] = rv2.pdf(arr2_2["x"])

def plots(a, b):
    sns.histplot(x=a, stat="density")
    sns.lineplot(data=b, x="x", y="PDF")
    plt.xlim(-5, 5)
    return plt.tight_layout()

plots(arr2, arr2_2)
```

```python
arr5 = variab(5)
E = expect (arr5, 1000)

rv2 = norm(E, 3)
arr5_2 = pd.DataFrame({"x": nums})
arr5_2["PDF"] = rv2.pdf(arr5_2["x"])

plots(arr5, arr5_2)
```

```python
arr14 = variab(14)
E = expect (arr14, 1000)

rv2 = norm(E, 3)
arr14_2 = pd.DataFrame({"x": nums})
arr14_2["PDF"] = rv2.pdf(arr14_2["x"])

plots(arr14, arr14_2)
```

```python
arr100 = variab(100)
E = expect (arr100, 1000)

rv2 = norm(E, 3)
arr100_2 = pd.DataFrame({"x": nums})
arr100_2["PDF"] = rv2.pdf(arr100_2["x"])

plots(arr100, arr100_2)
```
